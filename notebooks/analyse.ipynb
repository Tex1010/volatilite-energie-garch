```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n = 1000  # nombre de jours

# simulation de rendements journaliers (style énergie)
returns = np.random.normal(0, 0.015, n)

# génération d'un prix à partir d'un prix initial
price = [70]  # prix pétrole initial fictif

for r in returns:
    price.append(price[-1] * np.exp(r))

price = price[1:]

df = pd.DataFrame({
    "price": price
})

df.to_csv("../data/energy_data.csv", index=False)

df.head()
```

```python
plt.figure()
plt.plot(df["price"])
plt.title("Simulation prix énergie (pétrole)")
plt.show()
```
